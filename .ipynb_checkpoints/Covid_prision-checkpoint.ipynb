```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Cleaning up BOP

```python
clean_bop = bop.drop(columns = ['scrapedate',"lastupdated", "stateagency", "fac_state", 'match',
                    "bjs_fac_operator", "bjs_fac_linked", "bjs_fac_linkTexast", 'fac_address',
                    'fac_facilitytype', 'sum_inc_pop_bopmanaged', 'sum_inc_population_bopmanaged'], axis = 1)
```

```python
print(clean_bop.shape)
clean_bop.head()
```

```python
clean_bop = clean_bop.drop(clean_bop.loc[:,'bjs_fac_state':], axis =  1)
clean_bop.head()
```

```python
clean_bop['facilitytype'].value_counts()
```

```python
clean_bop['gender'] = bop['bjs_fac_gender']
clean_bop['security_lv'] = bop['bjs_fac_security']
clean_bop.head(1)
```

```python
clean_bop.dropna(thresh=11).isna().sum()
```

```python
columns = ['scrapedate','lastupdated','facility','sum_inc_tested', 'sum_inc_positive', 'sum_inc_deaths', 'sum_staff_positive', 'sum_staff_deaths',
           'fac_inc_positive', 'fac_inc_tested', 'fac_inc_negative', 'fac_inc_recovered', 'fac_inc_active',
           'fac_inc_pending', 'fac_staff_positive', 'fac_staff_recovered', 'fac_staff_active', 'fac_inc_restricted',
           'fac_inc_isolated', 'sum_staff_tested' ,'sum_inc_deathsconfirmed' , 'sum_inc_deathspresumed', 'sum_inc_deathspending',
           'bjs_fac_gender', 'bjs_fac_security', 'bjs_fac_operator', 'bjs_fac_malepop', 'bjs_fac_femalepop', 'bjs_fac_city',
           'bjs_fac_zip1', 'match']

new_texas = texas[columns]
print(new_texas.shape)
new_texas.head()
```

```python
new_texas['bjs_fac_gender'].value_counts().sum()
```

```python
bjs_texas = new_texas[new_texas['match'] == 'in BJS (matched)']
bjs_texas
```

```python
print(bjs_texas.isna().sum())
```

```python
new_texas.isna().sum()
```

```python
cols_to_drop = ['fac_inc_tested', 'fac_inc_negative', 'sum_inc_deathsconfirmed', 'sum_inc_deathspresumed',
                'sum_inc_deathspending', 'bjs_fac_femalepop', 'fac_inc_pending']

new_texas.drop(cols_to_drop, axis = 1, inplace = True)
bjs_texas.drop(cols_to_drop, axis = 1, inplace = True)

print(new_texas.shape)
print(bjs_texas.shape)
```

```python
new_texas.isna().sum()
```

```python
bjs_texas.dropna(thresh = 10).shape
```

```python
new_texas.dropna(thresh = 5).shape
```

```python
bjs_facilities = bjs_texas.groupby('facility').mean().index
```

```python
new_texas_facilities = new_texas.groupby('facility').mean().index
```

```python
missing_from_bjs = set(new_texas_facilities) - set(bjs_facilities)
```

```python
for fac in missing_from_bjs:
    print(new_texas[new_texas['facility'] == fac].isna().sum())
    print('---'*20)
```

```python
bjs_texas['case_fatality'] = bjs_texas['sum_inc_deaths']/bjs_texas['sum_inc_positive']
bjs_texas
```

```python
drop = list(bjs_texas[bjs_texas['sum_inc_deaths'].isna()].index)
bjs_texas.drop(drop, inplace = True)
```

## US-States-csv

```python
states = pd.read_csv('./data/us-states.csv')
big_t = states[states['state'] == 'Texas']
big_t
```

```python
big_t['case_fatality'] = big_t['deaths']/big_t['cases']
big_t.head()
```

```python
big_t['date'] = pd.to_datetime(big_t['date'])
```

```python
plt.plot(big_t['case_fatality']);
```

```python
plt.figure(figsize=(10, 7))
sns.lineplot(data = big_t, y = 'case_fatality', x = 'date' )
plt.xticks(['2020-02-28','2020-08-31','2021-02-01','2021-08-31'])
plt.title('Case Fatality over 2020-2021')
plt.show()
```
