````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier
from sklearn.linear_model import LogisticRegression, RidgeClassifierCV
from sklearn import metrics
````

1. Age | Objective Feature | age | int (days)
1. Height | Objective Feature | height | int (cm) |
1. Weight | Objective Feature | weight | float (kg) |
1. Gender | Objective Feature | gender | 1 - women, 2 - men |
1. Systolic blood pressure | Examination Feature | ap_hi | int |
1. Diastolic blood pressure | Examination Feature | ap_lo | int |
1. Cholesterol | Examination Feature | cholesterol | 1: normal, 2: above normal, 3: well above normal |
1. Glucose | Examination Feature | gluc | 1: normal, 2: above normal, 3: well above normal |

# Cleaning

````python
cardio = pd.read_csv('./data/cardio_train.csv', sep=';')
cardio
````

````python
# Dropping the Id
cardio.drop('id', axis = 1, inplace = True)
cardio.head()
````

````python
# converting age from days --> years
cardio['age'] = round(cardio['age'] / 365.25)
````

````python
cardio['age'] = cardio['age'].astype(int)
cardio.head()
````

````python
# Converting height from cm --> inches
cardio['height'] = cardio ["height"]*0.3937008
cardio.head(3)
````

````python
# Converting weight from kg --> lb 
cardio['weight'] = round(cardio['weight']* 2.20462, 1)
cardio.head(3)
````

````python
# Making cholesterol and glucose objects for classification
cardio['cholesterol'] = cardio['cholesterol'].map({1:'normal', 2:'above normal', 3: 'well above normal'})
cardio['gluc'] = cardio['gluc'].map({1:'normal', 2:'above normal', 3: 'well above normal'})
cardio.head()
````

````python
# rename ap_hi to systolic_bp AND ap_lo
cardio.rename(columns={'ap_hi' : 'systolic_bp ', 'ap_lo': 'diastolic_bp'}, inplace=True)
cardio.head()
````

````python
cardio['gender'] = cardio['gender'].map({1:'female', 2: 'male'})
````

````python
# Making them into age groups
age_group = []
for age in cardio['age']:
    if age <= 40:
        age_group.append('30 - 40')
    elif age > 40 and age <= 50:
        age_group.append('41 - 50')
    elif age > 50 and age <=60:
        age_group.append('51 - 60')
    else:
        age_group.append('60+')
        
cardio['age'] = age_group
cardio
````

````python
heart = pd.read_csv('./data/heart.csv')
heart.shape
````

````python
# Check for Na's
heart.isna().sum()
````

### Just checking Correlation of certain columns

````python
corr = heart.corr()

plt.figure(figsize = (3, 5))
sns.heatmap(corr[['HeartDisease']].sort_values(by = 'HeartDisease',ascending = False),
            annot = True)
plt.yticks(rotation = 45)
plt.show()
````

### Make `ExerciseAngina` and `Sex` Binary
**Sex:** Female = 1; Male = 0  
**ExerciseAngina:** Y = 1; N = 0

````python
heart['Sex'] = (heart['Sex'] == 'F') * 1
heart['ExerciseAngina'] = (heart['ExerciseAngina'] == 'Y') * 1
heart.head(3)
````

### Dummify Columns

````python
dummy_columns = ['ChestPainType', 'RestingECG', 'ST_Slope']
dummy_heart = pd.get_dummies(data = heart, columns=dummy_columns)
dummy_heart.head(3)
````

````python
# Will drop specific columns based on relevance since there are only 3 columns to drop
drop_from_dummy = ['ST_Slope_Flat', 'RestingECG_Normal', 'ChestPainType_TA']
dummy_heart.drop(drop_from_dummy, axis = 1, inplace = True)
````

## Getting ready for modeling (No gridsearch or feature engineering yet)

````python
X = dummy_heart.drop('HeartDisease', axis = 1)
y = dummy_heart['HeartDisease']
````

````python
# Check that its fairly well split
y.value_counts(normalize=True)
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=321, test_size = 0.2)
sc = StandardScaler()

Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)
````

### Guess we start with basic Logistics

````python
# Checking average Accuracy
cross_val_score(LogisticRegression(), Z_train, y_train)
````

````python
lg = LogisticRegression()
lg.fit(Z_train, y_train)
print('Training Accuracy:',lg.score(Z_train, y_train))
print('Testing Accuracy :',lg.score(Z_test, y_test))
````

````python
metrics.plot_confusion_matrix(lg, Z_test, y_test, cmap = 'magma');
````
